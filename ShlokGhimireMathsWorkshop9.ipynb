The derivative formulas used are:

- Forward difference: `(f(a+h) - f(a)) / h`
- Backward difference: `(f(a) - f(a-h)) / h`
- Central difference: `(f(a+h) - f(a-h)) / (2*h)`


```python
import numpy as np
import matplotlib.pyplot as plt
```

This function uses the method chosen by the user: `forward`, `backward`, or `central`.

```python
def derivative(func, a, method='central', h=0.001):
    if method == 'forward':
        return (func(a + h) - func(a)) / h
    elif method == 'backward':
        return (func(a) - func(a - h)) / h
    elif method == 'central':
        return (func(a + h) - func(a - h)) / (2 * h)
    else:
        raise ValueError("method must be 'forward', 'backward', or 'central'")
```

#1


```python
f = np.cos
true_derivative = lambda x: -np.sin(x)

h_values = [0.1, 0.01, 0.001, 0.0001]
a_values = [0, 1, 2]

for a in a_values:
    print('a =', a)
    actual = true_derivative(a)

    for h in h_values:
        answer = derivative(f, a, method='forward', h=h)
        error = abs(answer - actual)
        print('h =', h, ' answer =', answer, ' error =', error)
```

#2

```python
f = np.cos
a = 1
h = 0.001
actual = -np.sin(a)

forward_answer = derivative(f, a, method='forward', h=h)
backward_answer = derivative(f, a, method='backward', h=h)
central_answer = derivative(f, a, method='central', h=h)

print('Actual value   =', actual)
print('Forward answer =', forward_answer, ' Error =', abs(forward_answer - actual))
print('Backward answer=', backward_answer, ' Error =', abs(backward_answer - actual))
print('Central answer =', central_answer, ' Error =', abs(central_answer - actual))
```

#3


```python
x = np.linspace(0, 5*np.pi, 100)

numerical = derivative(np.sin, x, method='central', h=0.001)
true_value = np.cos(x)

plt.plot(x, numerical, 'ro', markersize=3, label='Central Difference')
plt.plot(x, true_value, 'b-', label='True Value')
plt.title('Central Difference Derivative of y = sin(x)')
plt.xlabel('x')
plt.ylabel('Derivative')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def derivative(f, x, h=0.001):
    return (f(x + h) - f(x - h)) / (2 * h)

x = np.linspace(0, 3, 100)

numerical = derivative(np.exp, x)
true_value = np.exp(x)

plt.plot(x, numerical, 'ro', markersize=3, label='Central Difference')
plt.plot(x, true_value, 'b-', label='True Value')

plt.title('Central Difference Derivative of y = e^x')
plt.xlabel('x')
plt.ylabel('Derivative')
plt.grid(True)
plt.legend()
plt.show()
```
