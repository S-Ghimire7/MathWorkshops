Example 1: Create and Understand Linear Systems. We will start with a simple problem. Consider the following linear systems:

a. x+3y=5
  
   𝑥−2𝑦=−5

  𝑥+𝑦=3

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5,5,100)
y1 = (5-x)/3
y2 = (5+x)/2
y3 = 3-x
fig, ax = plt.subplots()
ax.plot(x,y1, color='r')
ax.plot(x,y3, color='b')
ax.plot(x,y2, color='g')
ax.text(0.5,0.6, '$x+3y = 5$', color='r')
ax.text(-3,0.5,'$x-2y = -5$', color='g')
ax.text(1.2,2, '$x+y = 3$', color='b')
ax.set_xlim(-4,4)
ax.set_ylim(-2,6)
ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)
ax.set_xticks(list(range(-4,5)))
ax.set_aspect('equal')
ax.grid(True,ls=':')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

5# Create a 4x4 matrix containing random numbers in the range of 1 to 20. Have the user prompted for a number, check that the imputed value is a number and then check if the number is in the matrix. If the number is in the matrix, return the number of times that number is in the matrix.

```python
import numpy as np
mat = np.random.randint(1, 21, size=(4, 4))
print(mat)
inp = input("Enter a number : ")
if inp.isdigit():
    n = int(inp)
    cnt = np.count_nonzero(mat == n)
    if cnt > 0:
        print(f"The number {n} is found {cnt} time(s) in the matrix.")
    else:
        print(f"The number {n} is not found in the matrix.")
else:
    print("Invalid input!")
```

6# Create a matrix 𝑨 of order 5 × 5 with random entries. Calculate the rank of the matrix 𝑨 to check the number of linearly independent columns or rows and to check if the matrix is invertible. What are the other ways to check if the matrix is invertible?

```python
import numpy as np
mat_A = np.random.randint(1, 10, (5, 5))
print("Matrix A:")
print(mat_A)
rank_mat_A = np.linalg.matrix_rank(mat_A)
print("Rank of matrix A:", rank_mat_A)
if rank_mat_A == 5:
    print("Matrix is invertible")
else:
    print("Matrix is not invertible")
```

7# Generate a matrix 𝑨 of order 5 × 5 with random entries and generate a vector (column) 𝒙 of length 5 with random entries. Calculate Matrix-vector product 𝒃 = 𝑨𝒙. Now, we have a linear system, i.e. the matrix 𝑨 and a vector 𝒃. Solve this Linear System.

```python
import numpy as np
mat_A = np.random.randint(1, 10, (5, 5))
vec_x = np.random.randint(1, 10, 5)
print("Matrix A:")
print(mat_A)
print("Vector x:")
print(vec_x)
vec_b = np.dot(mat_A, vec_x)
print("Vector b = A x:")
print(vec_b)
x_solution = np.linalg.solve(mat_A, vec_b)
print("Solution of Ax = b:", x_solution)
```

9# Consider the following 10 points in X-Y plane: (−3, −7.6), (−2,1), (−1, 4.8), (0, 5), (1, 2.8), (2, −0.6), (3, −4), (4, −6.2), (5, −6), (7, 6.4). Plot these points in X-Y plane. Connect the points to guess the polynomial. Once you guess your polynomial, formulate a Linear System and then solve it to find the coefficient of the polynomial. Finally, check if the polynomial fits along with the above given points.

```python
import numpy as np
import matplotlib.pyplot as plt
data_x = np.array([-3, -2, -1, 0, 1, 2, 3, 4, 5, 7])
data_y = np.array([-7.6, 1, 4.8, 5, 2.8, -0.6, -4, -6.2, -6, 6.4])
poly_coeff = np.polyfit(data_x, data_y, 9)
x_smooth = np.linspace(min(data_x), max(data_x), 200)
y_smooth = np.polyval(poly_coeff, x_smooth)
plt.scatter(data_x, data_y, color='red')
plt.plot(x_smooth, y_smooth)
plt.title("Polynomial Fit")
plt.show()
y_fitted = np.polyval(poly_coeff, data_x)
print("Original y:", data_y)
print("Fitted y:  ", y_fitted)
```
