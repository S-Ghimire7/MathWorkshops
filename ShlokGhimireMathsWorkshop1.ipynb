2. Write a function myPow(x, n) that returns 𝑥 power 𝑛, where and n is a non-negative integer.
Do not use the ** operator or the math.pow function - use one while loop

```python
def myPow(x,n):
  while n != 0:
    power = x*x
    n -= 1
  return power


num = int(input('Enter a number: '))
exp = int(input('What power would you like to give the number? '))
myPow(num,exp)
```

3. Write a function SumDigits(n) that displays the sum of digits of “n”. Hint:
SumDigits(234) will be 2+3+4=9

```python
def SumDigits(n):
    total = 0
    while n != 0:
        r = n % 10
        total += r
        n //= 10
    print(total)

num = 234
SumDigits(num)
```

4. Write a function squaredSum that returns the sum of the squares of an array/list of
integers such that squaredSum([2,3,4]) gives 29. Write your solution in pure Python
to show the logic in your solution, do not use any library functions.

```python
def squaredSum(ar):
  total = 0
  for i in range(len(ar)):
    total += ar[i]**2
  print(total)

numbers = [2,3,4]
squaredSum(numbers)
```
